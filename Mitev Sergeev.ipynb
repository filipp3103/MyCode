## Import all necessary libriaries

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import cloudscraper
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

import math

from itunes_app_scraper.scraper import AppStoreScraper
```

```python
scraper = cloudscraper.create_scraper(delay=10,   browser={'custom': 'ScraperBot/1.0',})
```

```python
url = "https://giuliofrey.eu/datasources/"

r = scraper.get(url)
```

```python
soup = BeautifulSoup(r.text,'lxml')
```

```python
table = soup.find('table', class_='dataframe')
```

### Create a dataframe

```python
df = pd.DataFrame(columns=['App id', 'Rank', 'Name','Developer','Price','Genre', 'Average score', 'Number of reviews', 
                           'Approximate downloads'])
```

### Scrape the first site

```python
for row in table.tbody.find_all('tr'):    
    columns = row.find_all('td')
    
    if(columns != []):
        app_id = columns[0].text.strip()
        rank = columns[1].text.strip()
        name = columns[2].text.strip()

        df = df.append({'App id': app_id,  'Rank': rank, 'Name': name}, ignore_index=True)
```

```python
df
```

### Scrape the second source

```python
scraper = AppStoreScraper()
```

```python
for i in df.index:
    app_id = df.at[i,'App id']
    try:
        app_info = scraper.get_app_details(app_id, country="us")
        df.at[i,'Developer'] = app_info['sellerName']
        df.at[i,'Price'] = app_info['price']
        df.at[i,'Genre'] = app_info['primaryGenreName']
        df.at[i, 'Average score'] = app_info['averageUserRating']
        df.at[i, 'Number of reviews'] = app_info['userRatingCount']
        df.at[i, 'Approximate downloads'] = app_info['userRatingCount']*20
    except Exception:
        continue
```

```python
df = df.dropna()
df
```

```python
df['Price'] = df['Price'].astype(float)
df['Average score'] = df['Average score'].astype(float)
df['Number of reviews'] = df['Number of reviews'].astype(int)
df['Approximate downloads'] = df['Approximate downloads'].astype(int)
```

### Aggregate the genres

```python
df.Genre.value_counts()
```

```python
mapping = {'Education': ['Education', 'Reference'],
    'Games': ['Games'],
    'Entertainment & Music': ['Entertainment', 'Music'],
    'Business and Productivity': ['Business', 'Productivity', 'Utilities'],
    'Health': ['Lifestyle' 'Health & Fitness', 'Medical'],
    'Navigation': ['Travel', 'Navigation', 'Weather'],
    'Photo & Video': ['Photo & Video'],
    'Social Networking': ['Social Networking'],
    'Sports': ['Sports'],
    'Shopping': ['Shopping'],
    'Food & Drink': ['Food & Drink'],
    'Finance': ['Finance'],
    'News': ['News']
          }
```

```python
def get_macro_category(genre):
    for macro_category, genres in mapping.items():
        if genre in genres:
            return macro_category
    return 'Other'

df['Macro Category'] = df['Genre'].apply(get_macro_category)
df['Macro Category'].value_counts()
```

```python
category_counts = df['Macro Category'].value_counts()
small_categories = category_counts[category_counts < 10].index.tolist()

df['Macro Category'] = df['Macro Category'].apply(
    lambda x: 'Outside Good' if x in small_categories else x
)
df['Macro Category'].value_counts()
```
