```python
films = soup.findAll('div',class_='styles_root__ti07r')
films
```

```python
data = []
for p in range (1, 6):
    url = "https://www.kinopoisk.ru/lists/movies/top250/?page={p}"
    r = requests.get(url)
    films = soup.findAll('div',class_='styles_root__ti07r')

    for film in films:
        link = "https://www.kinopoisk.ru" + film.find('a', class_='base-movie-main-info_link__YwtP1').get('href')
        russian_name = film.find('a', class_='base-movie-main-info_link__YwtP1').find('span', class_ = 'styles_mainTitle__IFQyZ styles_activeMovieTittle__kJdJj').text
        original_name = film.find('span', class_ = 'desktop-list-main-info_secondaryText__M_aus').text
        country_and_type = film.find('a', class_='base-movie-main-info_link__YwtP1').find('span', class_ = 'desktop-list-main-info_truncatedText__IMQRP').text
        rate = film.find('span', class_ = 'styles_kinopoiskValuePositive__vOb2E styles_kinopoiskValue__9qXjg').text
    
        data.append([link, russian_name, original_name, country_and_type, rate])
```

```python
header = ["link", "russian_name", "original_name", "country_and_type", "rate"]
```

```python
df = pd.DataFrame(data, columns = header)
df 
```
